```python
import matplotlib.pyplot as plt 
import numpy as np 
from scipy import misc

face = misc.face(gray=True)
print(face.shape)
plt.imshow(face , cmap = plt.cm.gray )
plt.show()
```

```python
# Zoom 

# Hint : utiliser les dimisenions de la photo , slicing pour le zoom 
m , n = face.shape
zoom_face = face[m//4:-m//4 , n//4:-n//4]
zoom_face[zoom_face > 150] = 255
plt.imshow(zoom_face)
plt.show()

# Augment la limonisitée :




```

```python

# Comprésion du image 
zoom_face = zoom_face[::2 , ::2]
plt.imshow(zoom_face)
```
